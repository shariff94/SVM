# Pre-Processing

```python
data = data.drop(["ID"],axis=1)
```

```python
cat_cols = data.select_dtypes("category").columns
num_cols = data.select_dtypes(exclude="category").columns
print("Numerical Colums:",list(num_cols))
print("Categorical Colums:",list(cat_cols))
```

## There are NAs

```python
X = data.drop(["target"],axis=1)
y = data.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
cat_cols = cat_cols.drop("target")
```

```python
X_train[cat_cols].isna().sum()
```

```python
X_train[num_cols].isna().sum()
```

# Simple Imputer for the Categorical Columns

```python
from sklearn.impute import SimpleImputer
```

```python
cat_imputer = SimpleImputer(strategy="most_frequent")
num_imputer = SimpleImputer()
```

```python
cat_imputer.fit(X_train[cat_cols])
num_imputer.fit(X_train[num_cols])
```

```python
cat_df = pd.DataFrame(cat_imputer.transform(X_train[cat_cols]),columns=cat_cols)
```

```python
cat_df.head()
```

# Using KNN imputer for numerical

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
```

```python
knn_imputer = KNNImputer()
```

```python
knn_imputer.fit(X_train[num_cols])
```

```python
num_df = pd.DataFrame(knn_imputer.transform(X_train[num_cols]),columns=num_cols)
```

```python
num_df.head()
```

```python
X_imputed_train = pd.concat([cat_df,num_df],axis=1)[X_train.columns]
```

```python
X_imputed_test = pd.concat([pd.DataFrame(cat_imputer.transform(X_test[cat_cols]),columns=cat_cols),
                            pd.DataFrame(knn_imputer.transform(X_test[num_cols]),columns=num_cols)],axis=1)[X_train.columns]
```

```python
X_imputed_dummified_train = pd.get_dummies(X_imputed_train,drop_first=True)
X_imputed_dummified_test = pd.get_dummies(X_imputed_test,drop_first=True)
```

```python
# Scaling the numerical data of train
scaler = StandardScaler()
scaler.fit(X_imputed_dummified_train[num_cols])
X_imputed_dummified_train[num_cols] = scaler.transform(X_imputed_dummified_train[num_cols])
```

```python
# Scaling the test data
X_imputed_dummified_test[num_cols] = scaler.transform(X_imputed_dummified_test[num_cols])
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {"C":range(1,10),"kernel":['linear', 'poly', 'rbf', 'sigmoid'],'degree':range(2,10)}
```

```python
svc_model = SVC(class_weight="balanced")
grid_cv = GridSearchCV(svc_model,param_grid,verbose=15,n_jobs=-1)
```

```python
grid_cv.fit(X_imputed_dummified_train,y_train)
```

```python
grid_cv.best_estimator_
```

```python
y_train_pred = grid_cv.predict(X_imputed_dummified_train)
y_test_pred = grid_cv.predict(X_imputed_dummified_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_train,y_train_pred))
```

```python
print(classification_report(y_test,y_test_pred))
```
